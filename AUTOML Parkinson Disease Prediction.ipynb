```python
import numpy as np 
import pandas as pd 
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data=pd.read_csv('/kaggle/input/parkinsons-data-set/parkinsons.data')
data.head()
```

```python
def info(data):
    print(data.shape)
    print(data.size)
    
def missing(data):
    a,b = data.shape
    x = data.size
    if (a*b == x):
        print('No missing values! Go on exploring the data!')
print(info(data))
print(missing(data))
```

```python
features = data.drop(columns=['status','name'])
target   = data[['status']]
print(features.head())
print(target.head())
```

```python
X=features
y=target
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,target,train_size=0.80, test_size=0.20)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
tpot = TPOTClassifier(verbosity=2, max_time_mins=10)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
```
