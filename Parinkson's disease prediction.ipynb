Importing necessary libraries and packages

```python
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

Fetch the data from kaggle input folder, sice i have used kaggle notebooks and the data was taken from kaggle website

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Load the data and check the first 5 rows of the dataset given.

```python
data=pd.read_csv('/kaggle/input/parkinsons-data-set/parkinsons.data')
data.head()
```

Define two functions. One for getting the size and shape of data and other for checking if there is any missing values.

```python
def info(data):
    print(data.shape)
    print(data.size)
    
def missing(data):
    a,b = data.shape
    x = data.size
    if (a*b == x):
        print('No missing values! Go on exploring the data!')
```

```python
info(data)
missing(data)
```

Take the 'status' column as the target variable as it is a dependent feature, whereas all features are independent features. Also, make sure to eliminate the'names' column as it doesnot contribute for adding accuracy to our model.

```python
features = data.drop(columns=['status','name'])
target   = data[['status']]
print(features.head())
print(target.head())
```

Check for the best correlation b/w features and target variable.

```python
data.corr()
```

Do feature scaling to improve the performance of our model.

```python
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(features)
y=target
```

Use eXtreme Gradient Boosting Classifier for making predictions.

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=7)
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
def output(y):
    y= accuracy_score(y_test, y_pred)*100
    print('The accuracy of your model is:',y)
def cm(x):
    x = confusion_matrix(y_test,y_pred)
    print('The confusion matrix is:',x)
```

```python
print(output(y),cm(x))
```
